


__Keyword Arguments__

Keyword arguments, also known as __default arguments__, __named arguments__, or __optional arguments__. are a mechanism for providing a default value for one or more parameters.

They are incredibly prevalent and important, because they enable functions to provide a simple interface while maintaining configurability.

As an example, we've seen that we can call int to convert text to an equivalent integer, like `int("1101") # => 1101 (i.e. one thousand, one hundred, and one)`. However, the `int` callable also makes use of a default argument - a `base` in which to perform the conversion, which defaults to 10. So, in Python we can use the expression `int("1101", base=2) # => 13 (i.e. thirteen)`. The `int` callable presents a simple interface - there is only one required argument - but gives the caller additional configurability through the `base` parameter.

````python
def append_4(lst):
    lst.append(4)

x = []
append_4(x)
append_4(x)

x
````

````python
def parrot(voltage, state='a stiff', action='voom', type='Norwegian Blue'):
    print("-- this parrot wouldn't", action, end=' ')
    print('if you put', voltage, 'volts through it.')
    print('-- Lovely plumage, the', type)
    print("-- It's", state, "!")


parrot(100)

# In this function case we have defined few parameters while creating the function, so for the cases where we have defined some values, we are not required to define values while calling the function.

# Non-keyword argument after a keyword argument.
parrot(voltage=5.0, 'dead')
````

````python
# Duplicate value for the same argument.
parrot(110, voltage=220)
````

````python

# Unknown keyword argument.
parrot(actor='John Cleese')
````

````python
def ask_yn(promp, retries=4, complaint='Enter Y/N'):
    for i in range(retries):
        ok = input(promp).strip()
        if ok in ('Y' or 'y'):
            return True
        if ok in ('N', 'n'):
            return False
        print(complaint)
    return False

# Call with only the mandatory argument
ask_yn('Really quit?')
````

````python
# Call with one keyword argument
ask_yn('Ok to overwrite the file ?', retries=3)
````

````python
# Call with one keyword argument - in any order!
ask_yn('Update status?', complaint='Just Y/N')
````

````python

# Call with all of the keyword arguments
ask_yn('Send text?', retries=2, complaint='Y/N please!')
````
